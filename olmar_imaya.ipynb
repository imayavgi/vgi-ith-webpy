```python
%%zipline --start=2004-1-1 --end=2008-12-31
import sys
import logbook
import numpy as np

from zipline.finance import commission

zipline_logging = logbook.NestedSetup([
    logbook.NullHandler(),
    logbook.StreamHandler(sys.stdout, level=logbook.INFO),
    logbook.StreamHandler(sys.stderr, level=logbook.ERROR),
])
zipline_logging.push_application()

#STOCKS = ['AMD', 'CERN', 'COST', 'DELL', 'GPS', 'INTC', 'MMM']
STOCKS = ['AAPL', 'MSFT', 'XOM', 'JNJ', 'AMZN', 'FB', 'GE', 'T', 
        'BRK.B', 'JPM', 'GOOGL', 'PG', 'GOOG', 'VZ', 'PFE', 'WFC', 
        'CVX', 'HD', 'KO', 'INTC', 'CMCSA', 'MRK', 'PEP', 'PM', 'CSCO', 
        'BAC', 'V', 'DIS', 'IBM', 'UNH', 'MO', 'AMGN', 'C', 'ORCL', 'BMY', 'MDT', 
        'SLB', 'WMT', 'MMM', 'ABBV', 'GILD', 'MCD', 'AGN', 'CVS', 'MA', 'QCOM', 'HON', 
        'CELG', 'SBUX', 'UTX', 'LLY', 'BA', 'UNP', 'UPS', 'NKE', 'COST', 'LOW', 'ACN', 'SPG', 
        'TXN', 'WBA', 'USB', 'MDLZ', 'PCLN', 'CL', 'LMT', 'ABT', 'BIIB', 'TMO', 'GS', 'AIG', 'DD', 
        'TWX', 'DOW', 'AVGO', 'NEE', 'DUK', 'CB', 'OXY', 'EMC', 'TJX', 'AXP', 'CRM', 'KHC', 'COP', 
        'SO', 'F', 'AMT', 'ADBE', 'DHR', 'CAT', 'ESRX', 'D', 'MET', 'KMB', 'MON', 'BLK', 'EOG', 'TGT', 
        'GM', 'MCK', 'MS', 'GIS', 'BK', 'GD', 'RAI', 'RTN', 'PNC', 'PYPL', 'ADP', 'FDX', 'AET', 'NOC', 
        'NFLX', 'FOXA', 'HAL', 'BDX', 'ITW', 'KMI', 'SYK', 'YUM', 'EMR', 'PSX', 'PSA', 'HPE', 'CTSH', 'CME', 
        'ANTM', 'EXC', 'COF', 'MMC', 'SCHW', 'AEP', 'TRV', 'YHOO', 'PRU', 'PX', 'REGN', 'CI', 'EBAY', 'BSX', 
        'CCI', 'KR', 'SPGI', 'PCG', 'ICE', 'ECL', 'BBT', 'AFL', 'DAL', 'JCI', 'APD', 'STZ', 'ETN', 'NVDA', 'ALXN', 
        'AMAT', 'PLD', 'AON', 'INTU', 'HCN', 'ORLY', 'PPG', 'APC', 'WM', 'CAH', 'SYY', 'CSX', 'DG', 'NSC', 'PXD', 
        'SRE', 'ISRG', 'BAX', 'ADM', 'ZBH', 'STT', 'EQIX', 'VTR', 'HUM', 'LYB', 'ALL', 'PPL', 'AVB', 'EIX', 'EQR', 
        'ROST', 'SE', 'VLO', 'FISV', 'ILMN', 'WY', 'ED', 'AZO', 'EW', 'VRTX', 'HPQ', 'NWL', 'GLW', 'LUV', 'STJ', 
        'SHW', 'DFS', 'NEM', 'PEG', 'SYF', 'HCA', 'DE', 'ZTS', 'EA', 'DLTR', 'MNST', 'XEL', 'TSN', 'FIS', 'BXP', 
        'CBS', 'TEL', 'STI', 'BHI', 'MPC', 'K', 'VFC', 'EL', 'PCAR', 'AAL', 'WEC', 'CCL', 'ATVI', 'CAG', 'DVN', 
        'MYL', 'APA', 'ADI', 'OMC', 'CMI', 'NLSN', 'TYC', 'VNO', 'PAYX', 'PGR', 'IP', 'TAP', 'CERN', 'LB', 'GGP', 
        'DLPH', 'ES', 'SJM', 'O', 'HCP', 'DPS', 'APH', 'SWK', 'MTB', 'MCO', 'TROW', 'DTE', 'ROP', 'CTL', 'WLTW', 
        'IR', 'NUE', 'CLX', 'BRK.A', 'MJN', 'VMC', 'WMB', 'BCR', 'ULTA', 'FCX', 'CXO', 'AMP', 'A', 'EFX', 'VIAB', 
        'UAL', 'HIG', 'ABC', 'HSY', 'PH', 'DLR', 'NBL', 'ESS', 'GPC', 'XRAY', 'ROK', 'FE', 'HSIC', 'TDG', 'LVLT', 
        'AWK', 'WHR', 'FITB', 'ETR', 'NTRS', 'FTV', 'LLTC', 'LRCX', 'LH', 'MU', 'HES', 'AA', 'CTXS', 'EXPE', 'MAR', 
        'ADS', 'RHT', 'KIM', 'WDC', 'ADSK', 'BEN', 'MHK', 'HST', 'VRSK', 'HOT', 'PRGO', 'XLNX', 'WAT', 'MLM', 'AEE', 
        'KEY', 'DVA', 'CMS', 'CHD', 'PFG', 'EQT', 'SWKS', 'SYMC', 'AAP', 'CMG', 'FAST', 'GWW', 'BLL', 'TSCO', 'DGX', 
        'NOV', 'IVZ', 'RSG', 'RCL', 'MAS', 'FRT', 'CNC', 'COH', 'MCHP', 'SLG', 'CFG', 'KLAC', 'MKC', 'LLL', 'UHS', 
        'RF', 'MRO', 'PNR', 'AYI', 'GPN', 'COG', 'XEC', 'MAT', 'KMX', 'MAC', 'CPB', 'L', 'DOV', 'CINF', 'M', 'MSI', 
        'COL', 'AME', 'DHI', 'WRK', 'HRS', 'EXR', 'SCG', 'IFF', 'LKQ', 'TXT', 'LNC', 'KSU', 'CNP', 'CA', 'HRL', 
        'HBI', 'UDR', 'CHRW', 'BF.B', 'XL', 'IRM', 'WU', 'XRX', 'WFM', 'FBHS', 'EMN', 'HOG', 'STX', 'BBY', 'ALB', 
        'MOS', 'OKE', 'HOLX', 'SEE', 'CTAS', 'IPG', 'LNT', 'SNA', 'TSO', 'KORS', 'HAS', 'EXPD', 'VAR', 'AKAM', 
        'PNW', 'AJG', 'TSS', 'LEN', 'XYL', 'NFX', 'CBG', 'ALK', 'FFIV', 'VRSN', 'NI', 'PVH', 'NDAQ', 'AES', 'FL',
        'QRVO', 'JNPR', 'TRIP', 'WYNN', 'WYN', 'UNM', 'CMA', 'AMG', 'SRCL', 'KSS', 'GT', 'NTAP', 'HBAN', 'FLR', 
        'JBHT', 'MNK', 'BWA', 'AIV', 'UA', 'TIF', 'LEG', 'URI', 'DRI', 'ETFC', 'BBBY', 'AVY', 'ALLE', 'TMK', 'SWN', 
        'SIG', 'RRC', 'HP', 'PHM', 'UA.C', 'JEC', 'FMC', 'GRMN', 'NWSA', 'SNI', 'FLS', 'PKI', 'LUK', 'FTR', 'SPLS', 
        'GPS', 'DISCK', 'HAR', 'FTI', 'CF', 'ZION', 'JWN', 'RL', 'HRB', 'AIZ', 'RHI', 'TGNA', 'PBCT', 'DNB', 'NAVI',
        'FLIR', 'MUR', 'NRG', 'PDCO', 'PWR', 'RIG', 'AN', 'DISCA', 'CSRA', 'TDC', 'PBI', 'LM', 'FSLR', 'ENDP', 'R', 
        'CHK', 'OI', 'URBN', 'FOX', 'DO']

# On-Line Portfolio Moving Average Reversion

# More info can be found in the corresponding paper:
# http://icml.cc/2012/papers/168.pdf
def initialize(context, eps=1, window_length=5):
    context.stocks = STOCKS
    vsym = []
    for symbol in context.stocks :
        try:
            vsym.list.append( context.symbol(symbol) )    
        except:
            print('')
            print(" Unable to get symbol for %s " % symbol)
            
    context.sids = [context.symbol(symbol) for symbol in vsym ]   
    context.m = len(context.stocks)
    context.price = {}
    context.b_t = np.ones(context.m) / context.m
    context.last_desired_port = np.ones(context.m) / context.m
    context.eps = eps
    context.init = True
    context.days = 0
    context.window_length = window_length

    context.set_commission(commission.PerShare(cost=0))


def handle_data(algo, data):
    algo.days += 1
    if algo.days < algo.window_length:
        return

    if algo.init:
        rebalance_portfolio(algo, data, algo.b_t)
        algo.init = False
        return

    m = algo.m

    x_tilde = np.zeros(m)
    b = np.zeros(m)

    # find relative moving average price for each asset
    mavgs = data.history(algo.sids, 'price', algo.window_length, '1d').mean()
    for i, sid in enumerate(algo.sids):
        price = data.current(sid, "price")
        # Relative mean deviation
        x_tilde[i] = mavgs[sid] / price

    ###########################
    # Inside of OLMAR (algo 2)
    x_bar = x_tilde.mean()

    # market relative deviation
    mark_rel_dev = x_tilde - x_bar

    # Expected return with current portfolio
    exp_return = np.dot(algo.b_t, x_tilde)
    weight = algo.eps - exp_return
    variability = (np.linalg.norm(mark_rel_dev)) ** 2

    # test for divide-by-zero case
    if variability == 0.0:
        step_size = 0
    else:
        step_size = max(0, weight / variability)

    b = algo.b_t + step_size * mark_rel_dev
    b_norm = simplex_projection(b)
    np.testing.assert_almost_equal(b_norm.sum(), 1)

    rebalance_portfolio(algo, data, b_norm)

    # update portfolio
    algo.b_t = b_norm


def rebalance_portfolio(algo, data, desired_port):
    # rebalance portfolio
    desired_amount = np.zeros_like(desired_port)
    current_amount = np.zeros_like(desired_port)
    prices = np.zeros_like(desired_port)

    if algo.init:
        positions_value = algo.portfolio.starting_cash
    else:
        positions_value = algo.portfolio.positions_value + \
            algo.portfolio.cash

    for i, sid in enumerate(algo.sids):
        current_amount[i] = algo.portfolio.positions[sid].amount
        prices[i] = data.current(sid, "price")

    desired_amount = np.round(desired_port * positions_value / prices)

    algo.last_desired_port = desired_port
    diff_amount = desired_amount - current_amount

    for i, sid in enumerate(algo.sids):
        algo.order(sid, diff_amount[i])


def simplex_projection(v, b=1):
    v = np.asarray(v)
    p = len(v)

    # Sort v into u in descending order
    v = (v > 0) * v
    u = np.sort(v)[::-1]
    sv = np.cumsum(u)

    rho = np.where(u > (sv - b) / np.arange(1, p + 1))[0][-1]
    theta = np.max([0, (sv[rho] - b) / (rho + 1)])
    w = (v - theta)
    w[w < 0] = 0
    return w


# Note: this function can be removed if running
# this algorithm on quantopian.com
def analyze(context=None, results=None):
    import matplotlib.pyplot as plt
    fig = plt.figure()
    ax = fig.add_subplot(111)
    results.portfolio_value.plot(ax=ax)
    ax.set_ylabel('Portfolio value (USD)')
    plt.show()

```
